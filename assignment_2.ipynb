```python
import numpy as np
import math
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

def inclined(A1,A2,w):
  x = np.linspace(0, 10, 1000)
  fig = plt.figure()
  ax = plt.axes()
  k=0
  w1=math.radians(w)
  wcos=round(math.cos(w1),1)
  wsin=round(math.sin(w1),1)
  l1=A1[0]
  m1=A1[1]
  l2=A2[0]
  m2=A2[1]
  n1=-1*l1/m1                                             #slope of line1 in transformed axis
  n2=-1*l2/m2                                             #slope of line2 in transformed axis
  s1=((n1*wsin)/(1+n1*wcos))                              #slope of line1 represented in cartesian co-ordinate
  s2=((n2*wsin)/(1+n2*wcos))                              #slope of line2 represented in cartesian co-ordinate
  if (round((m1/l1)+(m2/l2),1)==2*wcos):
    if s1==s2:
      print('Slopes are equal mean lines are equally inclined')
      print('Slopes in X-Y Coordinates are : ', s1,s2)
      print('Figure is plotted in XY axes where w=90')
  else:
    print('HEY',wcos,s1,s2,((m1/l1)+(m2/l2)))
  plt.plot(x, n1*x +4, linestyle='solid')                 #Plots x+0.707y+4 =0
  plt.plot(x, n2*x +6, linestyle='dashed')                #Plots x+1.414y+4 =0

 

inclined([1,0.707],[2,1.414],45)
```
